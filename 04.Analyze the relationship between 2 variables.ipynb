### Analyze relationship between 2 variables

```python
# Read data from file
import pandas as pd
df=pd.read_csv('baseball.csv')
df.tail()
```

```python
def analyse_relationship(x,y): #x,y are pandas series
    correlation=x.corr(y)
    print('Correlation between '+x.name+' and '+y.name+' is ',correlation)
    
    if correlation>=0.7: print('Correlation is close to 1 indicates strong positive relationship')
    elif correlation<=-0.7: print('Correlation is close to -1 indicates strong negative relationship')
    elif abs(correlation)<0.7 and abs(correlation)>0.4: print('This is a weak relationship')
    else: print('Correlation is not close to -1 or 1 indicates weak to no relationship')
    
    import matplotlib.pyplot as plt
    plt.plot(x,y,'.')
    plt.xlabel(x.name)
    plt.ylabel(y.name)
    plt.title('Relationship between '+x.name+' and '+y.name)
```

```python
analyse_relationship(df['OnBasePercentage'],df['RunsScored'])
```

```python
analyse_relationship(df['Wins'],df['RunsScored'])
```

```python
import matplotlib.pyplot as plt
AL=df[df['League']=='AL']
NL=df[df['League']=='NL']
plt.figure()
plt.xlabel('On Base Percentage')
plt.ylabel('Runs Scored')
plt.scatter(AL['OnBasePercentage'], AL['RunsScored'], s=10, c='red', label='AL teams')
plt.scatter(NL['OnBasePercentage'], NL['RunsScored'], s=10, c='blue', label='NL teams')
plt.legend()
```
